```python
from glove import Glove
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
glove = Glove()
glove = glove.load("pos_glove_100d.model")
```

```python
def tsne_plot(model):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in model.dictionary.keys():
        tokens.append(glove.word_vectors[model.dictionary[word]])
        labels.append(word)
    
    tsne_model = TSNE(n_components=3, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

```python
tsne_plot(glove)
```
